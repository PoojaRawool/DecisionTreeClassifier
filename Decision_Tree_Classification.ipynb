```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

# Used the default settings of the DecisionTreeClassifier to train the model and check the classification metrics.

```python
model=tree.DecisionTreeClassifier()
model=model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.get_depth()
```

```python
model.get_n_leaves()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree.export import export_text
r = export_text(model)
print(r)
```

Looking at the training and testing accuracy score we can say that the model is not overfitting. Also with the default settings max_depth is 7.
